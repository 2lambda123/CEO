```python
import numpy as np
import math
from ceo import ceo
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
from IPython.html import widgets
from IPython.html.widgets import interact
from IPython.display import display
```

```python
def parametric_equation(coordinates,directions,s):
    x = coordinates[:,0] + s*directions[:,0]    
    y = coordinates[:,1] + s*directions[:,1]    
    z = coordinates[:,2] + s*directions[:,2]
    return x,y,z
```

```python
def rayTrace(radius=0.2,n_radius=3,n_theta=8,vc=-0.5,surf_0x=0.0,surf_0y=0.0,surf_0z=0.0,view='3d'):
    c = math.pi/180.0
    surf_O=[surf_0x*c,surf_0y*c,surf_0z*c]
    surf_0=[0.0,0.0,0.0]
    src = ceo.Source("R")
    rays = ceo.Bundle(src,radius,n_radius,n_theta,[0.0,0.0,-1])
    rays_origin = rays.coordinates.host()
    surf = ceo.Conic(vc,0.0,origin=surf_0,euler_angles=surf_O)
    tid = ceo.StopWatch()
    tid.tic()
    ceo.Transform_to_S(rays,surf)
    ceo.Intersect(rays,surf)
    ceo.Reflect(rays)
    ceo.Transform_to_R(rays,surf)
    tid.toc()
    if vc==0:
        s = 1
    else:
        s = -0.5/vc
    x,y,z = parametric_equation(rays.coordinates.host(),rays.directions.host(),s)
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    rays_current = rays.coordinates.host();
    for k in range(rays_origin.shape[0]):
        xp = [rays_origin[k,0],rays_current[k,0],x[k]]
        yp = [rays_origin[k,1],rays_current[k,1],y[k]]
        zp = [rays_origin[k,2],rays_current[k,2],z[k]]
        ax.plot3D(xp,yp,zp)
    ax.set_xlabel('X') 
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    if view=='top':
        ax.view_init(elev=90,azim=90)
    if view=='front':
        ax.view_init(elev=0,azim=90)
    if view=='side':
        ax.view_init(elev=0,azim=0)
```
