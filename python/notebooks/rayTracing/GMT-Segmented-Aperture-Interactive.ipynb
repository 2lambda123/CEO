```python
import numpy as np
import math
import ceo
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
from IPython.html import widgets
from IPython.html.widgets import interact
from IPython.display import display
```

```python
#def disp(x): print np.array_str(x,precision=1,suppress_small=True)
tid = ceo.StopWatch()
```

```python
nLenslet = 30
n = 8
nPx = n*nLenslet+1
onAxis = ceo.Source("R",zenith=0*60*math.pi/180/3600,azimuth=0*math.pi/4,resolution=(nPx,nPx))
src = ceo.Source("R",zenith=0*60*math.pi/180/3600,azimuth=0*math.pi/4,resolution=(nPx,nPx))
D = 25.5
n_radius = nPx
raysOnAxis = ceo.Bundle(onAxis,D,n_radius,64,[0.0,0.0,-25],type='box')
rays = ceo.Bundle(src,D,n_radius,64,[0.0,0.0,-25],type='box')
imgr = ceo.ShackHartmann(1, nPx-1, D ,DFT_osf=8,N_PX_IMAGE=4*nPx)
wfs = ceo.ShackHartmann(nLenslet, n, D/nLenslet)
```

```python
M1 = ceo.GMT_M1(D,n_radius)
M2 = ceo.GMT_M2(D,n_radius)
```

```python
def optical_system(x,_rays_):
    M2.blocking(_rays_)
    M1.trace(_rays_)
    M2.trace(_rays_)
    #rays.to_z_plane(5.83001949259)
    _rays_.to_sphere([0.0,0.0,x],1)
```

```python
def opd_fun(x):
    raysOnAxis.reset()
    optical_system(x,raysOnAxis)
    opd = raysOnAxis.optical_path_difference.host()*1e9
    P = raysOnAxis.vignetting.host()
    return np.std(opd[P>0])
```

```python
from scipy.optimize import minimize_scalar
#print opd_fun(5.83001949259)
res = minimize_scalar(opd_fun, bounds=(5.825, 5.835), method='bounded')
print res.success
print "Minimum wfe rms focus @ z=%6.4fm with rms wfe=%6.4fnm"%(res.x,opd_fun(res.x))
xyz0 = raysOnAxis.coordinates.host()
raysOnAxis.wavefront = onAxis
imgr0 = ceo.Imaging(1, nPx-1 ,DFT_osf=8,N_PX_IMAGE=75)
imgr0.propagate(onAxis)
wfs.calibrate(onAxis,0.5)
imgr.calibrate(onAxis,0.0)
```

```python
def ray_trace(mirror='M1',T=[0,0,0],O=[0,0,0],segId=0):
    
    print "IMAGER pixel scale  : %5.2fmas\n"%(1e3*imgr.pixel_scale_arcsec)
    print "IMAGER field-of-view: %5.2fmas\n"%(1e3*imgr.pixel_scale_arcsec*4*nPx)
    print "WFS pixel scale     : %5.2fmas\n"%(1e3*wfs.pixel_scale_arcsec)
    print "WFS field-of-view   : %5.2fmas\n"%(1e3*wfs.pixel_scale_arcsec*n)
    
    if mirror=="M1":
        M1.update(origin=T,euler_angles=O,idx=segId)
    if mirror=="M2":
        M2.update(origin=T,euler_angles=O,idx=segId)
    rays.reset()
    
    tid.tic()
    M2.blocking(rays)
    M1.trace(rays)
    M2.trace(rays)
    rays.to_z_plane(res.x)
    xyzf = rays.coordinates.host()
    rays.to_sphere(rays.chief_coordinates.host()[0,:],1)
    tid.toc()
    
    rays.wavefront = src    
        
    xf = xyzf[:,0]
    yf = xyzf[:,1]
    P = rays.vignetting.host()
    P1D = P.ravel(0)
    P = np.reshape(P,(n_radius,n_radius))
     
#    """
    ox = np.tan(src.zenith)*np.cos(src.azimuth)
    oy = np.tan(src.zenith)*np.sin(src.azimuth)
    u = linspace(-1,1,nPx)
    xx,yy = meshgrid(u,u)
    wfs_offset = P*(ox*xx + oy*yy)*D/2
    cu_wfs_offset = ceo.cuFloatArray(host_data=wfs_offset)
    src.phase = cu_wfs_offset
#    """
    
    imgr.reset()
    tid.tic()
    imgr.propagate(src)
    tid.toc()
    
    wfs.reset()
    wfs.analyze(src)

   # fig, ((ax1,ax2),(ax3,ax4)) = subplots(nrows=2, ncols=2)
    fig = figure()
    fig.set_size_inches(20,25)
    
    ax1 = subplot2grid( (3,2), (0,0))
    ax1.plot(xf[P1D>0]*1e6,yf[P1D>0]*1e6,'.')
    ax1.grid()
    ax1.set_aspect(aspect=1)
    ax1.set_xlabel('X [micron]')
    ax1.set_ylabel('Y [micron]')
    ax1.set_title('Spot diagram @ focus')
    
    #opd = rays.optical_path_difference.host()*1e9
    opd = src.phase.host(units='nm')
    ax2 = subplot2grid( (3,2), (0,1))
    hi = ax2.imshow(opd,interpolation='none')
    h = colorbar(hi,ax=ax2)
    h.set_label('OPD [nm]')
    grid()
    ax2.set_title( "WFE [nm]: RMS %4.2f & PtV %4.2f"%(np.std(opd[P>0]),(np.max(opd[P>0]) - np.min(opd[P>0]))))
    ax3 = subplot2grid( (3,2), (1,0))
    lim = 2*nPx*imgr.pixel_scale_arcsec
    h3 = ax3.imshow((imgr.frame.host()),interpolation='none',extent=[-lim,lim,-lim,lim],origin='lower')
    ax3.grid()
    ax3.set_xlabel('X [arcsec]')
    ax3.set_ylabel('Y [arcsec]')
    colorbar(h3,ax=ax3)
    
    ax4 = subplot2grid( (3,2), (1,1))
    h4 = ax4.imshow((wfs.frame.host()),interpolation='none')
    colorbar(h4,ax=ax4)
    
    ax5 = subplot2grid( (3,2), (2,0),colspan=2)
    h5 = imshow(1e3*wfs.c.host(units='arcsec').reshape(2*nLenslet
                                ,nLenslet).transpose(),interpolation='none')
    hc5 = colorbar(h5,ax=ax5)
    hc5.set_label('centroids [mas]')
    
    fig.tight_layout()
```
