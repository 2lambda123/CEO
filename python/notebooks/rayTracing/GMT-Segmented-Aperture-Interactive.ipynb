```python
def parametric_equation(coordinates,directions,zp):
    s = (zp - coordinates[:,2])/directions[:,2]
    x = coordinates[:,0] + s*directions[:,0]    
    y = coordinates[:,1] + s*directions[:,1]    
    z = zp*np.ones(x.shape)
    return x,y,z,s
```

```python
src = ceo.Source("R",zenith=0*math.pi/180/3600,azimuth=0*math.pi)
D = 25.5
n_radius = 500
rays = ceo.Bundle(src,D,n_radius,64,[0.0,0.0,-25],type='box')
print rays.chief_coordinates.host()
```

```python
M1 = ceo.GMT_M1(D,n_radius)
M2 = ceo.GMT_M2(D,n_radius)
```

```python
def ray_trace(mirror='M1',T=[0,0,0],O=[0,0,0],segId=0):
    if mirror=="M1":
        M1.update(origin=T,euler_angles=O,idx=segId)
    if mirror=="M2":
        M2.update(origin=T,euler_angles=O,idx=segId)
    rays.reset()
 
    tid.tic()
    M2.blocking(rays)
    M1.trace(rays)
    M2.trace(rays)
    rays.to_z_plane(5.83001949259)
    tid.toc()
    
    xyzf = rays.coordinates.host()
    xf = xyzf[:,0]
    yf = xyzf[:,1]
    P = rays.vignetting.host()
    P1D = P.ravel(0)
    fig, (ax1,ax2) = subplots(nrows=1, ncols=2)
    fig.set_size_inches(15,6)
    ax1.plot(xf[P1D>0]*1e6,yf[P1D>0]*1e6,'x')
    ax1.grid()
    ax1.set_aspect(aspect=1)
    ax1.set_xlabel('X [micron]')
    ax1.set_ylabel('Y [micron]')
    
    opl = rays.optical_path_length.host().T
    oa = (n_radius+1)*(n_radius/2)
    opl0 = 71.3549714274
    #opd = (P*rays.optical_path_difference.host())*1e9
    opd = P*rays.optical_path_difference.host()*1e9
    hi = ax2.imshow(np.reshape(opd,(n_radius,n_radius)),interpolation='none')
    h = colorbar(hi)
    h.set_label('OPD [nm]')
    grid()
    
    print "OPD RMS %4.2fnm"%np.std(opd[P>0])
```
