```python
import numpy as np
import math
import ceo
%pylab inline
from IPython.html import widgets
from IPython.html.widgets import interact
from IPython.display import display
```

```python
def zernike(zi=5):
    print "j=%d"%zi
    npx = 512;
    u = np.linspace(-1,1,npx)
    x,y = np.meshgrid(u,u)
    r = np.hypot(x,y)
    o = np.arctan2(y,x)
    nra = np.ceil((-3 + sqrt( 9 + 8*(zi-1) ))/2)
    print "n=%d"%nra
    n_mode = (nra+1)*(nra+2)/2.0
    #print n_mode
    a = np.zeros(n_mode)
    a[zi-1] = 1.0
    Z = ceo.ZernikeS(nra,a)
    cuo = ceo.cuDoubleArray(host_data=o)
    cur = ceo.cuDoubleArray(host_data=r)
    tid = ceo.StopWatch()
    tid.tic()
    S = Z.surface(cur,cuo)
    tid.toc()
    S_ = S.host()
    figure()
    imshow(S_,interpolation=None)
    colorbar()
```

```python
i = interact(zernike,zi=(1,66))
```

```python
def zernike_dx_coefs_1(j0,n0,m0):
    for np in range(0,n0):
        jp = np*(np+1)/2 + 1
        mp = np%2
        while (mp<=np):    
            if mp==0:
                b = ceo.Zern_dx_coef(j0,n0,m0,jp,np,mp)
                if b!=0:
                    print " ==> j=%2d,n=%d,m=%d,b=%4.2f"%(jp,np,mp, b)
                jp += 1
                mp += 2
            else:
                b = ceo.Zern_dx_coef(j0,n0,m0,jp,np,mp)
                if b!=0:
                    print " ==> j=%2d,n=%d,m=%d,b=%4.2f"%(jp,np,mp, b)
                jp += 1
                b = ceo.Zern_dx_coef(j0,n0,m0,jp,np,mp)
                if b!=0:
                    print " ==> j=%2d,n=%d,m=%d,b=%4.2f"%(jp,np,mp, b)
                jp += 1
                mp += 2            
```

```python
def zernike_dx_coefs(n):
    j = n*(n+1)/2 + 1
    m = n%2
    while (m<=n):    
        if m==0:
            print "j=%2d,n=%d,m=%d"%(j,n,m)
            zernike_dx_coefs_1(j,n,m)
            j += 1
            m += 2
        else:
            print "j=%2d,n=%d,m=%d"%(j,n,m)
            zernike_dx_coefs_1(j,n,m)
            j += 1
            print "j=%2d,n=%d,m=%d"%(j,n,m)
            zernike_dx_coefs_1(j,n,m)
            j += 1
            m += 2            
```
