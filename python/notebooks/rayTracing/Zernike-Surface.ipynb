```python
import numpy as np
import math
import ceo
%pylab inline
from IPython.html import widgets
from IPython.html.widgets import interact
from IPython.display import display
```

```python
def zernike(zi=5):
    print zi
    npx = 512;
    u = np.linspace(-1,1,npx)
    x,y = np.meshgrid(u,u)
    r = np.hypot(x,y)
    o = np.arctan2(y,x)
    nra = np.ceil((-3 + sqrt( 9 + 8*(zi-1) ))/2)
    print nra
    n_mode = (nra+1)*(nra+2)/2.0
    print n_mode
    a = np.zeros(n_mode)
    a[zi-1] = 1.0
    Z = ceo.ZernikeS(nra,a)
    cuo = ceo.cuDoubleArray(host_data=o)
    cur = ceo.cuDoubleArray(host_data=r)
    tid = ceo.StopWatch()
    tid.tic()
    S = Z.surface(cur,cuo)
    tid.toc()
    S_ = S.host()
    figure()
    imshow(S_,interpolation=None)
    colorbar()
```

```python
i = interact(zernike,zi=(1,6600))
```
