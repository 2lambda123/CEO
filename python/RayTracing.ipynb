```python
import numpy as np
import math
from ceo import ceo
%pylab inline
```

```python
src = ceo.Source("R")
```

```python
rays = ceo.Bundle(src,0.2,3,8,[0.0,0.0,-1])
```

```python
rays_origin = rays.coordinates.host()
def rays_display(rays):
    return None
    #print "RAYS:\n . coordinates:\n" , np.array_str(rays.coordinates.host().T,precision=3,suppress_small=True) ,  "\n . directions:\n",  np.array_str(rays.directions.host().T,precision=3,suppress_small=True)
rays_display(rays)
```

```python
surf = ceo.Conic(-0.5,0.0,origin=[0.4,0.0,0.0],euler_angles=[0.0,-0*math.pi/20,0*math.pi/3])
```

```python
def parametric_equation(rays,s):
    x = rays.coordinates.host()[:,0] + s*rays.directions.host()[:,0]    
    y = rays.coordinates.host()[:,1] + s*rays.directions.host()[:,1]    
    z = rays.coordinates.host()[:,2] + s*rays.directions.host()[:,2]
    return x,y,z
```

```python
x,y,z = parametric_equation(rays,1.2)
#print x
```
