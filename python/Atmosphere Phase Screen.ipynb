Phase screens are computed with an `Atmosphere` and a `Source` object.
Lets import the `ceo` module first.

```python
import math
import numpy as np
import plotly.plotly as ply
import plotly.graph_objs as plyg
import ceo
```

The atmosphere is composed of a single layer at 10km.

```python
atm = ceo.Atmosphere(0.15,30,altitude=10e3,wind_speed=10,wind_direction=0)
```

A V band on-axis source is defined first,

```python
n = 64
src = ceo.Source("V",resolution=(n,n))
```

and an off-axis source at a zenith angle such as the chief rays of both source are separated by $D$ meter at 10km

```python
D = 4.0
src_oa = ceo.Source("V",zenith=D/10e3,azimuth=0,resolution=(n,n))
```

The $n \times n$ phase screens of $L=8$m width each are computed next

```python
L = 8.0
p = L/n
atm.get_phase_screen(src,1,p,n,p,n,0.0)
atm.get_phase_screen(src_oa,1,p,n,p,n,0.0)
```

The two phase screens at 10km are plotted together

```python
trace = plyg.Heatmap( z=src.phase.host()*1e6 ,
                     x = np.linspace(0,8,n),
                     y = np.linspace(0,8,n),
                     colorscale="Portland",
                     zauto = False,
                    zmin=-5.5,zmax=1.5)
trace_oa = plyg.Heatmap( z=src_oa.phase.host()*1e6 ,
                     x = np.linspace(0,L,n),
                     y = np.linspace(D,D+L,n),
                     colorscale="Portland",
                     zauto = False,
                    zmin=-5.5,zmax=1.5)
data = plyg.Data([trace,trace_oa])
fig = plyg.Figure(data=data,layout=plyg.Layout(margin=plyg.Margin(l=50,r=0,t=50,b=50),
                                                autosize=False,height=500,width=300))
ply.iplot(fig)
```
