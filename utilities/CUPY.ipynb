```python
import numpy as np
import cupy as cp
import ceo
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gmt = ceo.GMT_MX()
src = ceo.Source('V',rays_box_size=25.5,rays_box_sampling=401,rays_origin=[0,0,25])
src>>(gmt,)
+src
src.wavefront.rms(-9)
```

```python
phase = ceo.ascupy(cp,src.phase)
```

```python
plt.imshow(phase.get()*1e9)
plt.colorbar()
```
