```python
import math
import numpy as np
import ceo
%pylab inline
```

```python
nPx = 256
D = 25.5
src    = ceo.Source("R",zenith=[0,0],azimuth=[0,0],
                    rays_box_size=D,rays_box_sampling=nPx,rays_origin=[0.0,0.0,25])
gmt = ceo.GMT_MX(D,nPx,N_SRC=src.N_SRC)
```

```python
src.reset()
gmt.propagate(src)
pm = gmt.M1.piston_mask
#print pm.shape
imshow(np.reshape(pm[0][1,:],(nPx,nPx)),interpolation='none')
```

```python
src.wavefront.piston()
```

```python
src.wavefront.piston(pm)
```

```python
PS = ceo.SegmentPistonSensor(gmt,src)
src.reset()
gmt.propagate(src)
PS.P = gmt.M1.piston_mask
print PS.P
```

```python
M = PS.M
R = D/2
rc = 4.387
rp = 7.543
fig,ax =subplots()
fig.set_size_inches((10,10))
ax.imshow(M[0].sum(axis=0).reshape((nPx,nPx)),extent=[-R,R,-R,R],interpolation='None',cmap='Blues',origin='lower')
#ax.imshow(M[7,:].reshape((n,n)),extent=[-R,R,-R,R],interpolation='None',cmap='Blues',origin='lower')
grid()
xticks([-rp,-rc,rc,rp])
yticks([-rp,-rc,rc,rp])
c1 = Circle((0,0),radius=rc, color='r', fill=False)
ax.add_patch(c1)
c2 = Circle((0,0),radius=rp, color='r', fill=False)
ax.add_patch(c2)
```

```python
fig,ax =subplots()
fig.set_size_inches((10,10))
P = gmt.M1.piston_mask[0].sum(axis=0)
P = src.wavefront.amplitude.host()[0,:]
ax.imshow(P.reshape(nPx,nPx) + np.logical_and(P,M[0].sum(axis=0)).reshape(nPx,nPx),
          extent=[-R,R,-R,R],interpolation=None,cmap='Blues',origin='lower')
grid()
xticks([-rp,-rc,0,rc,rp])
yticks([-rp,-rc,0,rc,rp])
c1 = Circle((0,0),radius=rc, color='r', fill=False)
ax.add_patch(c1)
c2 = Circle((0,0),radius=rp, color='r', fill=False)
ax.add_patch(c2)
```
