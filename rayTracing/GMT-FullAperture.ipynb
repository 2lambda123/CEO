### Ray bundle:

Polar coordinates ray bundle
* radius 12.5m
* 5 rays accros the radius
* 8 rays on the azimuths

```python
src = ceo.Source("R")
D = 25
nPx = 500
radius = D
n_radius = nPx
src  = ceo.Source("R",rays_box_size=D,rays_box_sampling=nPx,rays_origin=[0.0,0.0,-1.0])
rays_coord = src.rays.coordinates.host()
xr = rays_coord[:,0]
yr = rays_coord[:,1]
zr = rays_coord[:,2]
#print xr
#print yr
```

### Aperture

```python
src.reset()
A = ceo.Aperture(25.0,0.0,n_radius)
A.vignetting(src.rays)
P = src.rays.vignetting.host().T
imshow(np.reshape(P,(n_radius,n_radius)),interpolation='none')
colorbar()
```

### M1

GMT M1 definition:
* curvature radius: 36m
* conic constant: 1-0.9982857

```python
M1 = ceo.Conic(-1.0/36.0,1-0.9982857)
```

### Reflection from M1

```python
tid = ceo.StopWatch()
tid.tic()
ceo.Transform_to_S(src,M1)
print src.rays.chief_coordinates.host()
print src.rays.coordinates.host()[500**2/2,:]
ceo.Intersect(src,M1)
print src.rays.chief_coordinates.host()
print src.rays.coordinates.host()[500**2/2,:]
ceo.Reflect(src)
ceo.Transform_to_R(src,M1)
tid.toc()
```

### M2

GMT M1 definition:
* curvature radius: 4.163900922m
* conic constant: 1-0.71692784

```python
M2 = ceo.Conic(1.0/4.163901,1-0.7169273,origin=[0.0,0.0,-20.262476],euler_angles=[0.0,0.0*math.pi/180/3600,0.0])
```

### Reflection from M2

```python
tid.tic()
ceo.Transform_to_S(src,M2)
ceo.Intersect(src,M2)
ceo.Reflect(src)
ceo.Transform_to_R(src,M2)
src.rays.go_sphere(26.0924924672,1)
tid.toc()
coords = src.rays.coordinates.host()
xr = np.vstack((xr,coords[:,0]))
yr = np.vstack((yr,coords[:,1]))
zr = np.vstack((zr,coords[:,2]))
```
